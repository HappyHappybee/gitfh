```python
import pandas as pd
import glob
import re
from functools import reduce

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from collections import Counter

import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
import nltk
```

```python
nltk.download()
```

```python
all_files = glob.glob('8장_data/myCabinetExcelData*.xls')

all_files
```

```python
all_files_data = []

for file in all_files:
    data_frame = pd.read_excel(file)
    all_files_data.append(data_frame)

all_files_data[0]
```

```python
all_files_data_concat = pd.concat(all_files_data, axis=0, ignore_index=True)

all_files_data_concat
```

```python
all_files_data_concat.to_csv('8장_data/riss_bigdata.csv', encoding='utf-8', index = False)
```

```python
all_title = all_files_data_concat['제목']

all_title
```

```python
stopWords = set(stopwords.words("english"))
lemma = WordNetLemmatizer()
```

```python
words = []  

for title in all_title:
    EnWords = re.sub(r"[^a-zA-Z]+", " ", str(title))    
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
    words.append(EnWordsTokenStopLemma)
```

```python
print(words)
```

```python
words2 = list(reduce(lambda x, y: x+y,words))
print(words2)
```

```python
count = Counter(words2)

count
```

```python
word_count = dict()

for tag, counts in count.most_common(50):
    if(len(str(tag))>1):
        word_count[tag] = counts
        print("%s : %d" % (tag, counts))
```

```python
del word_count['big']
del word_count['data']
```

```python
plt.figure(figsize=(12, 5))
plt.xlabel("word")
plt.ylabel("count")
plt.grid(True)

# Check if word_count is a dictionary (assuming it has words as keys and counts as values)
if not isinstance(word_count, dict):
  print("word_count is not a dictionary. Please ensure it's a dictionary with words as keys and counts as values.")
  exit()  # Exit the program if not a dictionary

sorted_Keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)

plt.bar(range(len(word_count)), sorted_Values, align='center')

# Set rotation to a valid option (number of degrees, 'vertical', or 'horizontal')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation=85)  # Set rotation to 85 degrees (counter-clockwise)
# OR
# plt.xticks(range(len(word_count)), list(sorted_Keys), rotation='vertical')  # Set labels vertically

plt.show()
```

```python
all_files_data_concat['doc_count'] = 0
summary_year = all_files_data_concat.groupby('출판일', as_index=False)['doc_count'].count()
summary_year
```

```python
plt.figure(figsize=(12,5))
plt.xlabel("year")
plt.ylabel("doc-count")
plt.grid(True)

plt.plot(range(len(summary_year)), summary_year['doc_count'])
plt.xticks(range(len(summary_year)), [text for text in summary_year['출판일']])

plt.show()
```

```python
stopwords=set(STOPWORDS)
wc=WordCloud(background_color='ivory', stopwords=stopwords, width=800, height=600)
cloud=wc.generate_from_frequencies(word_count)

plt.figure(figsize=(8,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
